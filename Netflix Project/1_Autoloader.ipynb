## Incremental Data loading using Autoloader

```python
%sql
Create Schema netflix_catalog.net_schema;
```

```python
checkpoint_location = 'abfss://silver@slnetflixprojectdatalake.dfs.core.windows.net/checkpoint'
```

```python
df = spark.readStream\
    .format("cloudFiles")\
    .option("cloudFiles.format","csv")\
    .option("cloudFiles.schemaLocation",checkpoint_location)\
    .option("checkpointLocation",checkpoint_location)\
    .load("abfss://raw@slnetflixprojectdatalake.dfs.core.windows.net")
```

```python
display(df,checkpointLocation=checkpoint_location)
```

```python
df.writeStream\
    .option("checkpointLocation", checkpoint_location)\
    .trigger(processingTime = "10 seconds")\
    .start("abfss://bronze@slnetflixprojectdatalake.dfs.core.windows.net/netflix_titles")
```
